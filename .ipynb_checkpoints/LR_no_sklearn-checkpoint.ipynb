```python
class Custom_Lr:
    def __init__(self):
        self.m = None
        self.d = None
        
    def fit(self,X_train,y_train):

        num = 0
        den = 0

        for i in range(X_train.shape[0]):

            num = num + ((X_train.iloc[i] - X_train.mean())*(y_train.iloc[i] - y_train.mean()))           

            den = den + ((X_train.iloc[i] - X_train.mean())**2)

        self.m = num/den

        self.d = ( y_train.mean() - ( self.m * X_train.mean() ))
        
        print(self.m)
        print(self.d)
    
    def predict(self,X_test):
        print (X_test)
        return self.m * X_test + self.d
```
