### Building a class which is similar to the simple linear regression class in sklearn but from scratch 

```python
class MyCustomModel:
    def __init__(self):
        #based on "y = mx + b" principal
        self.m = None 
        self.b = None
    def fit(self,X_train,y_train):
        
        num = 0 # numarator
        den = 0 # denominator
        
        for i in range (X_train.shape[0]):
            num = num + ((X_train[i] - X_train.mean()) * (y_train[i] - y_train.mean()))
            den = den + ((X_train[i] - X_train.mean()) * (X_train[i] - X_train.mean()))
        self.m = num/den # calculating the value of "m"
        self.b = y_train.mean() - (self.m * X_train.mean()) # calculating the value of "b"
        print(self.m)
        print(self.b)
        
    def predict(self,X_test): # making prediction
        return self.m * X_test + self.b
```
